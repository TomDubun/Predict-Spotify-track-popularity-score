```python
french1 = data[data['genres'].str.contains("french")]
french2 = data[data['genres'].str.contains("francais")]
df_fr = pd.concat([french1,french2])
df_fr = df_fr.drop_duplicates(subset=['spotify_uri'])
df_fr
```

I am interested in the french music market. I think running a machine learning algorithm on all countries would not make sense since accuracy scores would be low. Music consumption is different from one country/region/culture to another.

```python
rap = df_fr[df_fr['genres'].str.contains("rap")]
hip_hop = df_fr[df_fr['genres'].str.contains("hip hop")]
df_hip_hop = pd.concat([hip_hop,rap])
df_hip_hop = df_hip_hop.drop_duplicates(subset=['spotify_uri'])
df_hip_hop['genre']='hip_hop'
df_hip_hop
```

From the french music market, I will explore main genres.

I am bundling rap and hip hop into the rap category.

```python
pop = df_fr[df_fr['genres'].str.contains("pop")]
variete = df_fr[df_fr['genres'].str.contains("variete")]
df_pop = pd.concat([pop,variete])
df_pop = df_pop.drop_duplicates(subset=['spotify_uri'])
df_pop['genre']='pop'
df_pop
```

I am above bundling pop and "variété" into pop music. Variété is a french category very close to pop music

```python
electro = df_fr[df_fr['genres'].str.contains("electro")]
house = df_fr[df_fr['genres'].str.contains("house")]
df_electro = pd.concat([electro,house])
df_electro = df_electro.drop_duplicates(subset=['spotify_uri'])
df_electro['genre']='electro'
df_electro
```

I am above bundling electro and house as electro music. 

```python
df_rock = df_fr[df_fr['genres'].str.contains("rock")]
df_rock['genre']='rock'
df_rock
```

The last category I will be focusing is rock.

```python
df = pd.concat([df_rock,df_hip_hop, df_electro, df_pop])
df = df.drop_duplicates(subset=['spotify_uri'])
df['spotify_uri'] = df['spotify_uri'].str.replace(r'spotify:artist:', '')
df = df.rename(columns={'name': 'artist_name', 'genre': 'artist_genre', 'spotify_uri': 'artist_id'})
df = df.drop(['Unnamed: 0','genres'], axis = 1)
df
```

## SPOTIFY FRANCE PLAYLISTS

```python
def get_user_playlists(user_id, n_playlists=-1):
    playlist_names, playlist_ids = [], []
    
    # fetch first 20 playlists 
    results = sp.user_playlists(user=user_id, limit=20)
    user_playlists = results['items']
    time.sleep(2)
    
    # fetch playlist 21 - end
    for ofs in range(20,n_playlists,20):
        results = sp.user_playlists(user=user_id, limit=20, offset=ofs)
        user_playlists += results['items']
        time.sleep(2)
        
    # option to limit the no of playlists retreived
    if n_playlists > 0:
        if n_playlists < results['total']:
            user_playlists = user_playlists[:n_playlists]
        
    # extract name and id
    for pl in user_playlists:
        playlist_ids.append(pl['id'])
        playlist_names.append(pl['name'])
        playlist_ids = playlist_ids[:n_playlists]
        playlist_names = playlist_names[:n_playlists]
        
    # overview df
    user_playlists_total = pd.DataFrame({'name': playlist_names, 'ids': playlist_ids})
    return playlist_ids, user_playlists_total
```

This function enables to get playlists ids for a given spotify user.
